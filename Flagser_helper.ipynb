# Random graphs

```python
def transform_to_flagser_file(name_in, name_out):
  res_file = []
  df = pd.read_csv(name_in, delimiter=' ', names=['n1', 'n2', 'w'])
  # Index starts from 0 
  df['n1'] = df['n1'] - 1
  df['n2'] = df['n2'] - 1
  # no loops
  df = df[df['n1']!=df['n2']]

  res_file.append('dim 0')
  # 0 for dim 0
  rep_0 = np.repeat(0, max(df['n1'].max(), df['n2'].max()) + 1)
  res_file.append(' '.join(map(str,rep_0)))

  res_file.append('dim 1')
  for item in df.values:
    res_file.append(' '.join(map(str,item)))
  np.savetxt(name_out, res_file, fmt='%s')
```

```python
transform_to_flagser_file('erdos_renyi_edges.txt', 'erdos_renyi_edges_flagser.txt')
transform_to_flagser_file('barabasi_albert_edges.txt', 'barabasi_albert_edges_flagser.txt')
transform_to_flagser_file('geometric_edges.txt', 'geometric_edges_flagser.txt')
transform_to_flagser_file('chung_lu_edges.txt', 'chung_lu_edges_flagser.txt')
transform_to_flagser_file('geom_chung_lu_edges.txt', 'geom_chung_lu_edges_flagser.txt')
```

# Diagrams

```python
def get_intervals_from_file_flagser_out(name, max_number):
  lst = [[], [], [], [], [], [], [], [], []] # H0, H1, H2, H3, H4, H5
  file1 = open(name, 'r')
  h = -1 #dimension
  while True:
      #read one line
      line = file1.readline()
      #The remaining homology groups are trivial. - stop condition
      #print(line)
      if not line:
          break
      if line == 'The remaining homology groups are trivial.\n':
          break
      elif '# persistence intervals in dimension' in line: 
        h = h+1
      elif '[' in line:
        #extra symbols
        interval = line.replace('[', '').replace(')', '').replace(' ', '').replace('\n', '').split(',')
        #first and second number
        if interval[1] == '':
          lst[h].append([int(interval[0]), max_number])
        else:
          lst[h].append([int(interval[0]), int(interval[1])])

  file1.close
  return lst
```

## Betti curves

```python
def get_betti_numbers(intervals, max_number):
    betti = []
    for i in range(0,max_number):
        sum = 0
        for interval in intervals:
            if (interval[0] <= i) and (interval[1] > i):
                sum = sum + 1
        betti.append(sum)
    return betti
```
